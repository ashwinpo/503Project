```r
library(tidyverse)
library(superml)
```

```r
fake = read.csv("Fake.csv", encoding = "UTF-8")
true = read.csv("True.csv", encoding = "UTF-8")
```

```r
fake$label = rep("Fake", nrow(fake))
true$label = rep("True", nrow(true))
data = rbind(true, fake)
```

```r
## set the seed to make your partition reproducible
set.seed(503)
train_ind <- sample(seq_len(nrow(data)), size = floor(0.001 * nrow(data)))

train <- data[train_ind, ]
test <- data[-train_ind, ]
```

```r
# Ensure relatively balanced classes
train %>% group_by(label) %>% summarize(n = n())
test %>% group_by(label) %>% summarize(n = n())
```

```r
train = train %>% select(-text) %>% cbind(.,train_tf_features)
```

```r
# initialise the class
tfv <- TfIdfVectorizer$new(min_df = 0.1, max_df = 0.9, max_features = 800,
                           remove_stopwords = TRUE, ngram_range = c(1,3))

# we fit on train data
tfv$fit(train$text)

train_tf_features <- tfv$transform(train$text)
test_tf_features <- tfv$transform(test$text)

dim(train_tf_features)
dim(test_tf_features)
```
